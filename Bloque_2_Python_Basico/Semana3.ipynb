```python
# =============================================================================
# Script: Calculadora de Índice de Masa Corporal (IMC)
# Descripción: Calcula el IMC basado en el peso y la estatura del usuario
#              y proporciona una clasificación según los estándares de salud.
# =============================================================================

def calculadora_imc(peso, estatura):
    """
    Calcula el valor numérico del IMC.

    Args:
        peso (float): El peso del usuario en kilogramos.
        estatura (float): La estatura del usuario en metros.

    Returns:
        float: El índice de masa corporal calculado.
    """
    # Fórmula del IMC: peso dividido por la estatura al cuadrado
    imc = peso / (estatura ** 2)
    return imc

def clasificador_imc(imc):
    """
    Determina la categoría de salud según el valor del IMC y la imprime.

    Args:
        imc (float): El valor del índice de masa corporal a clasificar.
    """
    clasificacion = "Tu clasificación es "

    # Estructura condicional para determinar la categoría según rangos estándar
    if imc < 18.5:
        print(clasificacion + "bajo peso")
    elif imc < 25.0:
        print(clasificacion + "peso normal")
    elif imc < 30.0:
        print(clasificacion + "sobrepeso")
    elif imc < 35.0:
        print(clasificacion + "obesidad tipo I")
    elif imc < 40.0:
        print(clasificacion + "obesidad tipo II")
    else:
        # Para valores de 40.0 o superiores
        print(clasificacion + "obesidad tipo III")

# --- Bloque Principal de Ejecución ---

# Captura de datos del usuario convirtiendo la entrada de texto a flotante (decimal)
try:
    estatura = float(input("Introduce tu estatura en metros: "))
    peso = float(input("Introduce tu peso en kilogramos: "))

    # Llamada a la función de cálculo
    imc_resultado = calculadora_imc(peso, estatura)

    # Muestra el resultado formateado a dos decimales
    print(f"Tu IMC es: {imc_resultado:.2f}")

    # Clasifica y muestra el estado de salud basado en el resultado
    clasificador_imc(imc_resultado)

except ValueError:
    # Manejo de errores en caso de que el usuario no introduzca números válidos
    print("Error: Por favor, introduce valores numéricos válidos para peso y estatura.")
```

```python
# =============================================================================
# Script: Calculadora de IMC Multi-usuario
# Descripción: Calcula el IMC para múltiples personas en una sesión,
#              solicitando el nombre y permitiendo repetir el proceso.
# =============================================================================

def calculadora_imc(peso, estatura):
    """
    Calcula el valor numérico del IMC.

    Args:
        peso (float): El peso en kilogramos.
        estatura (float): La estatura en metros.

    Returns:
        float: El índice de masa corporal.
    """
    # Calcula el IMC (Peso / Estatura al cuadrado)
    return peso / (estatura ** 2)

def clasificador_imc(imc):
    """
    Determina la categoría de salud según el valor del IMC.

    Args:
        imc (float): El valor del IMC.
    """
    prefijo = "Tu clasificación es "

    # Lógica de rangos de salud
    if imc < 18.5:
        print(prefijo + "bajo peso")
    elif imc < 25.0:
        print(prefijo + "peso normal")
    elif imc < 30.0:
        print(prefijo + "sobrepeso")
    elif imc < 35.0:
        print(prefijo + "obesidad tipo I")
    elif imc < 40.0:
        print(prefijo + "obesidad tipo II")
    else:
        print(prefijo + "obesidad tipo III")

def ejecutar_calculadora():
    """
    Función principal que gestiona el bucle de entrada de datos y usuarios.
    """
    print("--- Bienvenido a la Calculadora de IMC ---")

    while True:
        try:
            # Solicitar datos del usuario actual
            nombre = input("\nIntroduce el nombre del usuario (o escribe 'salir' para terminar): ")

            # Opción para romper el bucle manualmente
            if nombre.lower() == 'salir':
                break

            estatura = float(input(f"Introduce la estatura de {nombre} en metros (ej: 1.75): "))
            peso = float(input(f"Introduce el peso de {nombre} en kilogramos (ej: 70): "))

            # Procesamiento de datos
            resultado_imc = calculadora_imc(peso, estatura)

            # Salida de resultados personalizada
            print("-" * 30)
            print(f"Resultados para: {nombre}")
            print(f"IMC: {resultado_imc:.2f}")
            clasificador_imc(resultado_imc)
            print("-" * 30)

        except ValueError:
            # Captura errores si el usuario no ingresa números válidos
            print("\n[Error]: Por favor, introduce valores numéricos válidos para peso y estatura.")
            continue

        # Preguntar si desea continuar con otro usuario
        opcion = input("\n¿Deseas calcular el IMC de otra persona? (s/n): ").lower()
        if opcion != 's':
            break

    print("\nGracias por usar la calculadora. ¡Hasta pronto!")

# Punto de entrada del script
if __name__ == "__main__":
    ejecutar_calculadora()
```
