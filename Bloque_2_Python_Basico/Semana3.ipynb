```python
# Calculadora de Índice de Masa Corporal (IMC)

def calculadora_imc(peso, estatura):
    imc = peso / (estatura ** 2)
    return imc

def clasificador_imc(imc):
    clasificacion = "Tu clasificación es "
    if imc < 18.5:
        print(clasificacion + "bajo peso")
    elif imc < 25.0:
        print(clasificacion + "peso normal")
    elif imc < 30.0:
        print(clasificacion + "sobrepeso")
    elif imc < 35.0:
        print(clasificacion + "obesidad tipo I")
    elif imc < 40.0:
        print(clasificacion + "obesidad tipo II")
    else:
        print(clasificacion + "obesidad tipo III")

estatura = float(input("Introduce tu estatura en metros: "))
peso = float(input("Introduce tu peso en kilogramos: "))

imc = calculadora_imc(peso, estatura)
print(f"Tu IMC es: {imc:.2f}")
clasificador_imc(imc)
```
