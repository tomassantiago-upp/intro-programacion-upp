```python
# Ejemplo de try/except

print("---Calculadora edad---")

while True:
    try:
    # Intentamos convertir la edad a un número
        edad_str = input("Ingresa tu edad: ")
        edad_num = int(edad_str)
        break

    except ValueError:
        # Si no es un número, muestra un error y vuelve al inicio
        print("ERROR: Dato inválido, favor de introducir un número.")

print("Tu edad dentro de 10 años es: ", edad_num + 10)

```

```python
# Ejemplo 2 de try/except

print("---DIVISOR DE GASTOS DE MI STARTUP---")
presupuesto = 50000

try:
    empleados_str = input("Ingrese el número de empleados: ")
    empleados_num = int(empleados_str)
    bono_por_persona = presupuesto / empleados_num
    print("A cada empleado le toca un bono de: ", bono_por_persona)

except ValueError:
    print("ERROR: dato inválido, favor de introducir un número")

except ZeroDivisionError:
    print("ERROR: no puedes dividir entre 0 empleados.")
```
