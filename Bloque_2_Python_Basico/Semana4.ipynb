```python
# Ejercicio 1

def verificar_usuario():
    current_users = ["monica08", "ian08", "athena07", "lea07", "michelle08"]
    new_users = ["ATHENA07", "yael08", "fred08", "Ian08", "hector08"]
    available = True

    for new_user in new_users:
        for current_user in current_users:
            if new_user.lower() == current_user.lower():
                available = False
                break
            else:
                available = True
        if available:
            print(new_user + " está disponible")
        else:
            print("El nombre "+ new_user + " no está disponible. Elegir otro")


if __name__ == "__main__":
    verificar_usuario()
```
