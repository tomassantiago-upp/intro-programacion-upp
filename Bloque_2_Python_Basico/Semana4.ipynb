```python
# Ejercicio 1 con complejidad O(n+m), más óptimo

def verificar_usuario():
    current_users = ["monica08", "ian08", "athena07", "lea07", "michelle08"]
    new_users = ["ATHENA07", "yael08", "fred08", "Ian08", "hector08"]

    # 1. Normalizamos a minúsculas y convertimos a set para búsqueda instantánea
    current_users_set = {user.lower() for user in current_users}
    new_users_set = {user.lower() for user in new_users}

    # 2. Encontramos la intersección (usuarios que ya existen)
    usuarios_tomados = current_users_set.intersection(new_users_set)

    # 3. Verificamos disponibilidad
    for user in new_users:
        if user.lower() in usuarios_tomados:
            print(f"El nombre {user} no está disponible. Elegir otro.")
        else:
            print(f"{user} está disponible.")

if __name__ == "__main__":
    verificar_usuario()
```
