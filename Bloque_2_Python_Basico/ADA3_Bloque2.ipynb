```python
import json
import os

ARCHIVO_BD = "BloqueII\Semana5\estudiantes\estudiantes.json"

def cargar_datos():
    if os.path.exists(ARCHIVO_BD):
        try:
            with open(ARCHIVO_BD, "r", encoding="utf-8") as archivo:
                return json.load(archivo)
        except json.decoder.JSONDecodeError:
            # Si el archivo existe pero está vacío o corrupto, devolvemos una lista vacía
            return []

    else:
        return []

def guardar_datos(lista_estudiantes):
    with open(ARCHIVO_BD, "w", encoding="utf-8") as archivo:
        json.dump(lista_estudiantes, archivo, indent=4, ensure_ascii=False)

def capturar_calificacion(mensaje):
    while True:
        try:
            calificacion = float(input(mensaje))
            if 0 <= calificacion <= 100:
                return calificacion
            else:
                print("¡La calificación debe estar en 0 y 100!")
        except ValueError:
            print("ERROR: Debes ingresar números, no letras")

def calcular_promedio(adas, proyecto):
    promedio_adas = sum(adas) / len(adas) if len(adas) > 0 else 0
    calificacion_final = (promedio_adas * 0.5) + (proyecto * 0.5)
    return round(calificacion_final, 2)

def iniciar_sistema():
    print("Iniciando sistema de captura de calificaciones")

    base_datos = cargar_datos()
    print(f"Registros actuales en memoria: {len(base_datos)}")

    while True:
        print("\n" + "-"*40)
        print("1. Captura nuevo estudiante")
        print("2. Ver base de datos")
        print("3. Salir y guardar")
        opcion = input("Selecciona una opción:")

        if opcion == "1":
            nombre = input("Ingresa el nombre del estudiante: ").strip().title()
            apellido = input("Ingresa el apellido del estudiante: ").strip().title()

            calificaciones_adas = []
            print("\n--- Captura de ADAS  ---")
            contador_ada = 1
            while True:
                calif = capturar_calificacion(f"Calificacion ADA {contador_ada}:")

                calificaciones_adas.append(calif)
                contador_ada += 1

                ans = input("¿Deseas agregar otra calificación? (y/n): ")
                if ans.lower() == 'n':
                    break

            print("\n --- Captura de Proyecto ---")
            calif_proyecto = capturar_calificacion("Calificación del proyecto final: ")

            promedio_final = calcular_promedio(calificaciones_adas, calif_proyecto)

            nuevo_estudiante = {
                "nombre": nombre,
                "apellido": apellido,
                "calificaciones_adas": calificaciones_adas,
                "proyecto_final": calif_proyecto,
                "promedio_final": promedio_final
            }
            base_datos.append(nuevo_estudiante)
            print(f"Estudiante {nombre} {apellido} guardado exitosamente.")

        elif opcion == "2":
            if len(base_datos) == 0:
                print("No hay estudiantes registrados en la base de datos.")
            else:
                print("\n --- LISTA DE ESTUDIANTES ---")
                for estudiante in base_datos:
                    print(f"{estudiante['nombre']} {estudiante['apellido']} ")
                    print(f"- Calificaciones ADAS: {estudiante['calificaciones_adas']}")
                    print(f"- Proyecto: {estudiante['proyecto_final']}")
                    print(f"- Promedio final: {estudiante['promedio_final']}")

        elif opcion == "3":
            guardar_datos(base_datos)
            print("Datos guardados en JSON exitosamente.")
            break

        else:
            print("Opción no válida")

if __name__ == "__main__":
    iniciar_sistema()
```
