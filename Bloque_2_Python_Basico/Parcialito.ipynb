```python
# ==========================================================
# CÓDIGO DAÑADO - NO FUNCIONA CORRECTAMENTE
# ==========================================================
# Instrucción para el equipo: Este código es lineal y frágil.
# Deben transformarlo en un sistema basado en FUNCIONES y JSON.

import json
import os

inventario = []

# --- PROBLEMA 1: Este bucle no está en una función ---
while True:
    print("\n--- INVENTARIO DE REFACCIONES ---")
    print("1. Añadir Pieza")
    print("2. Mostrar Todo")
    print("3. Salir")

    opcion = input("Selecciona: ")

    if opcion == "1":
        # --- PROBLEMA 2: No hay validación de errores (Try/Except) ---
        nombre = input("Nombre de la pieza: ")
        cantidad = int(input("Cantidad: "))
        precio = float(input("Precio unitario: "))

        pieza = {
            "nombre": nombre,
            "stock": cantidad,
            "costo": precio
        }
        inventario.append(pieza)
        print("Pieza añadida localmente.")

    elif opcion == "2":
        # --- PROBLEMA 3: No muestra el valor total (cantidad * precio) ---
        print("\nLista de Materiales:")
        for p in inventario:
            print(f"- {p['nombre']}: {p['stock']} unidades.")

    elif opcion == "3":
        # --- PROBLEMA 4: No guarda nada en JSON al salir ---
        print("Cerrando sistema...")
        break

# ¿Cómo harían para que al volver a abrir el programa los datos sigan ahí?
```
